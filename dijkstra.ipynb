```python


import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from math import inf


```

```python


wmat  = [[ 0 , 2 , 0 , 0 , 0 , 1 , 0 , 0 ],
        [ 2 , 0 , 2 , 2 , 4 , 0 , 0 , 0 ],
        [ 0 , 2 , 0 , 0 , 3 , 0 , 0 , 1 ],
        [ 0 , 2 , 0 , 0 , 4 , 3 , 0 , 0 ],
        [ 0 , 4 , 3 , 4 , 0 , 0 , 7 , 0 ],
        [ 1 , 0 , 0 , 3 , 0 , 0 , 5 , 0 ],
        [ 0 , 0 , 0 , 0 , 7 , 5 , 0 , 6 ],
        [ 0 , 0 , 1 , 0 , 0 , 0 , 6 , 0 ]]
```

```python
matriz=np.matrix(wmat)
```

```python
G2 = nx.from_numpy_matrix(matriz)
```

```python
nx.draw (G2,with_labels=True, node_size=2000, node_color="black", node_shape="p", alpha=0.5, linewidths=4, font_size=25, 
        font_color="white", font_weight="bold", width=2)
```

```python
def  find_all ( matriz , inicio , final = - 1 ):
    n  =  len ( matriz )
    dist  = [ inf ] * n
    dist [ inicio ] =  matriz [ inicio ] [ inicio ]   # 0
    spVertex  = [False ] * n
    padre  = [ - 1 ] * n
    ruta  = [{}] * n
    for  count  in range(n - 1):
        minix  =  inf
        u  =  0
        for v  in range(len(spVertex)):
            if spVertex[v] ==  False  and  dist[v] <=  minix :
                 minix  =  dist[v]
                 u  =  v
        spVertex[u] = True
        for v  in range(n):
            if not(spVertex[v]) and matriz[u][v] != 0 and dist[u] + matriz[u][v] < dist[v]:
                 padre[v] = u
                 dist[v] = dist[u] + matriz[u][v]
    for i in range(n):
        j = i
        s = []
        while padre[j] != -1:
               s.append(j)
               j = padre[j]
        s.append(inicio)
        ruta[i] = s[::-1]

    return (dist[final], ruta[final]) if final >= 0 else (dist, ruta)        
```
